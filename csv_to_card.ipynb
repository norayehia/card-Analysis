```python
import pandas as pd

# Load the data from CSV (assuming the data is in a file called 'cards.csv')
df = pd.read_csv('/content/drive/MyDrive/help/workong/frelanc working remote/lonivtyintime/share/cards/Healthy_and_Unhealthy_Cards_Extended_Levels.csv')
```

```python
# Function to generate card content
def generate_card_content(row):
    card_type = row['Card Type']
    card_name = row['Card Name']
    base_effect = row['Effect']

    # Starting with basic card content
    card_content = f"**Card Name:** {card_name}\n**Card Type:** {card_type}\n**Effect:** {base_effect}\n"

    # Loop through levels to generate level-based effects
    card_content += "\n**Level Effects:**\n"
    for level in range(1, 16):  # Assuming levels are 1 to 15
        level_col = f'Level {level}'
        if level_col in row:
            card_content += f"Level {level}: {row[level_col]}\n"

    return card_content
```

```python
# Loop through all rows to generate content for each card
card_contents = []
for index, row in df.iterrows():
    card_contents.append(generate_card_content(row))

# Output the generated card contents (or save to a file)
for content in card_contents:
    print(content)
    print("="*50)  # Separator between cards
```
